```python
#第二題完整
input_num = input().split(',')
n = int(input_num[0])
k = int(input_num[1])

num_list=list(range(1,n+1))
num_candidate_list=list(range(1,n+1))
input1 = input().split(',') 

#將list裡面的元素全部轉換成int型態
for i in range(len(input1)):
    input1[i] = int(input1[i])
#將代表城鎮 i 到各城鎮之距離放入list1之中    
list1 = []
for i in range(n):
    input2 = input().split(',')
    for j in range(len(input2)):
        input2[j] = int(input2[j])
    list1.append(input2)
    
totallength = 999999999999 #總長度
smallest_num=0

if k==1:
    
    for i in range(len(input1)):
        length = 0 
        for j in range(len(input1)):
            distance_for_j = input1[j]*list1[i][j]
            length += distance_for_j #計算出i城鎮的總長度

        if length < totallength: #找出最小總長度的城鎮
            totallength = length
            smallest_num = i+1 #找出該城鎮的編號

    tour_list=[]
    tour_list.append(smallest_num)
    num_candidate_list.remove(smallest_num)
    
    print(smallest_num,totallength,sep=";")
    
if k==2:
    
    for i in range(len(input1)):
        length = 0 
        for j in range(len(input1)):
            distance_for_j = input1[j]*list1[i][j]
            length += distance_for_j #計算出i城鎮的總長度

        if length < totallength: #找出最小總長度的城鎮
            totallength = length
            smallest_num = i+1 #找出該城鎮的編號

    tour_list=[]
    tour_list.append(smallest_num)
    num_candidate_list.remove(smallest_num)
    smallest_index = smallest_num - 1

    second_index = list1[smallest_index].index(max(list1[smallest_index]))
    second_num = second_index + 1        
    tour_list.append(second_num)
    num_candidate_list.remove(second_num)
    
    ans=0
    list4=[]
    for i in num_candidate_list:

        Small_distance_num_candidate_list=9999999999

        for j in tour_list:

            distance_num_candidate=list1[i-1][j-1]

            if distance_num_candidate < Small_distance_num_candidate_list:
                    Small_distance_num_candidate_list = distance_num_candidate

        list4.append(Small_distance_num_candidate_list)

    for i in range(len(num_candidate_list)):
        ans=ans + input1[(num_candidate_list[i]-1)]*list4[i]
    
    print(tour_list[0],tour_list[1],sep=",",end=";")
    print(ans)

else:
    
    for i in range(len(input1)):
        length = 0 
        for j in range(len(input1)):
            distance_for_j = input1[j]*list1[i][j]
            length += distance_for_j #計算出i城鎮的總長度

        if length < totallength: #找出最小總長度的城鎮
            totallength = length
            smallest_num = i+1 #找出該城鎮的編號

    tour_list=[]
    tour_list.append(smallest_num)
    num_candidate_list.remove(smallest_num)

    smallest_index = smallest_num - 1
    second_index = list1[smallest_index].index(max(list1[smallest_index]))
    second_num = second_index + 1        
    tour_list.append(second_num)
    num_candidate_list.remove(second_num)    

    while True:
        list3=[]
        for i in num_candidate_list:

            Small_distance_from_RemainTown_ToCandidate=9999999999

            for j in tour_list:

                distance_from_RemainTown_ToCandidate=list1[i-1][j-1]

                if distance_from_RemainTown_ToCandidate < Small_distance_from_RemainTown_ToCandidate:
                    Small_distance_from_RemainTown_ToCandidate = distance_from_RemainTown_ToCandidate

            list3.append(Small_distance_from_RemainTown_ToCandidate)

        now_furthest_distance = max(list3)
        third_num = num_candidate_list[list3.index(now_furthest_distance)]
        tour_list.append(third_num)
        num_candidate_list.remove(third_num)

        if len(tour_list)==k:
            break

    ans=0
    list4=[]
    for i in num_candidate_list:

        Small_distance_num_candidate_list=9999999999

        for j in tour_list:

            distance_num_candidate=list1[i-1][j-1]

            if distance_num_candidate < Small_distance_num_candidate_list:
                    Small_distance_num_candidate_list = distance_num_candidate

        list4.append(Small_distance_num_candidate_list)

    for i in range(len(num_candidate_list)):
        ans=ans + input1[(num_candidate_list[i]-1)]*list4[i]
        
    for i in tour_list:
        if i==tour_list[-1]:
            print(i,end=';')
        else:
            print(i,end=',')
    print(ans)
```
