```python
def importdata(path):
    data=[]
    with open(file=path, mode="r", encoding="utf-8") as f:
        reader = csv.reader(f, skipinitialspace=True)

        headers = next(reader)
        for line in reader :

            data.append(line)

        f.close()
        
    return data,headers
```

```python
def remove_comma(text):
    double_quotes_all=[]
    double_quotes_even=[]
    for i in range(len(text)):
        if text[i]=='\"':
            double_quotes_all.append(i)
    for i in range(0,len(double_quotes_all), 2):
        double_quotes_even.append([double_quotes_all[i],double_quotes_all[i+1]])

    after_remove=''
    for i in range(len(text)):
        whether=True

        for j in double_quotes_even:

            if text[i]==',' and  j[0]<=i<=j[1]:
                whether=False
                
            if text[i]=='\"':
                whether=False

        if whether:
            after_remove+=text[i]
    
    return after_remove
```

```python
class Submission:
    def __init__(self, user_id, status, score, time):
        self.user_id = user_id 
        self.status = status 
        self.score = score 
        self.time = time
        
```

```python
class Problem:
    def __init__(self, name,challenge):
        self.name = name 
        self.submissions = []
        self.challenge=challenge
    
    def add_submission(self, submission): 
        self.submissions.append(submission)
       
    def mission_a(self,time):
        time=datetime.datetime.strptime(time,"%Y/%m/%d-%H:%M:%S") 
        dict_user={}
        for i in self.submissions:
            if i.time<=time:
                
                if i.user_id in dict_user:
                    if dict_user[i.user_id][1]<i.time:
                        dict_user[i.user_id]=[i.status,i.time]
                
                else:
                    dict_user[i.user_id]=[i.status,i.time]
        ac=0
        not_ac=0
        for i in dict_user.keys():
            if dict_user[i][0]=='Accepted':
                ac+=1
            else:
                not_ac+=1

        return ac,not_ac 
    
    def ac_count(self):
        ac_count=0
        not_ac_count=0
        for i in self.submissions:
            
            if i.status=='Accepted':
                ac_count+=1
            else:
                not_ac_count+=1

        return ac_count,not_ac_count
```

```python
class Challenge:
    def __init__(self, name):
        self.name = name 
        self.problems = []
        
    def add_problem(self, problem): 
        self.problems.append(problem)
  
    def mission_b(self,time):
        
        ac_num=0
        not_ac_num=0
        
        for i in self.problems:
            ac,not_ac=i.mission_a(time)
            
            ac_num+=ac
            not_ac_num+=not_ac
        
        ans_ac=int(ac_num/len(self.problems))
        ans_not_ac=int(not_ac_num/len(self.problems))
        
        return ans_ac,ans_not_ac
        
        
    def mission_c(self,rate):
        
        smaller_than_rate=0
        
        for i in self.problems:
            ac_count,not_ac_count=i.ac_count()
            problem_rate=ac_count/(ac_count+not_ac_count)
            if problem_rate<=rate:
                smaller_than_rate+=1
                
        return smaller_than_rate
        
        
    def mission_d(self,number):   
        
        bigger_than_number=0
        
        for i in self.problems:
            ac_count,not_ac_count=i.ac_count()
            total_number=ac_count+not_ac_count
            if total_number>=number:
                bigger_than_number+=1
                
        
        return bigger_than_number

            
```

```python
def importdata(path):
    
    problem_list=[]
    problem_class_list=[]
    challenge_list=[]
    challenge_class_list=[]

    with open(file=path, mode="r", encoding="utf-8") as f:


        reader = csv.reader(f, skipinitialspace=True)

        headers = next(reader)


        for row in reader:

            time=datetime.datetime.strptime(row[9],"%Y-%m-%d %H:%M:%S")
            submission=Submission(row[2],row[1],int(row[5]), time )



            if row[4] not in problem_list:
                problem_list.append(row[4])
                problem=Problem(row[4],row[3])
                problem.add_submission(submission)
                problem_class_list.append(problem)
            else:
                for i in problem_class_list:
                    if i.name==row[4]:

                        i.add_submission(submission)


            if row[3] not in challenge_list:
                challenge_list.append(row[3])
                challenge=Challenge(row[3])
                challenge_class_list.append(challenge)


        f.close()

        
    return problem_list,problem_class_list,challenge_list,challenge_class_list
```

```python
problem_list,problem_class_list,challenge_list,challenge_class_list=importdata(path)

for i in challenge_class_list:
    for j in problem_class_list:
        if i.name==j.challenge:
            i.add_problem(j)
```

```python
def print_command(command_list):
    if command_list[0]=='a':
        
        if command_list[1] not in problem_list:
            print(-1)

        else:
            for i in problem_class_list:
                if i.name==command_list[1]:
                    ac,not_ac = i.mission_a(command_list[2])
                    print(ac,not_ac,sep=",")
                    break

    elif command_list[0]=='b':

        if command_list[1] not in challenge_list:
            print(-1)

        else:
            for i in challenge_class_list:
                if i.name==command_list[1]:
                    ans_ac,ans_not_ac=i.mission_b(command_list[2])
                    print(ans_ac,ans_not_ac,sep=',')
                    break

    elif command_list[0]=='c':

        if command_list[1] not in challenge_list:
            print(-1)

        else:
            for i in challenge_class_list:
                if i.name==command_list[1]:
                    smaller_than_rate=i.mission_c(float(command_list[2]))
                    print(smaller_than_rate)
                    break

    elif command_list[0]=='d':

        if command_list[1] not in challenge_list:
            print(-1)

        else:
            for i in challenge_class_list:
                if i.name==command_list[1]:
                    bigger_than_number=i.mission_d(int(command_list[2]))
                    print(bigger_than_number)
                    break
```
